Identity operators

```python
f = ['apple', 'banana', 'cherry']
g = ['toyato', 'mazda', 'honda']
i = f
print(f is i) #same object
print(f is g) #different objects
print(f is not i) #not same object
print(f is not g) #not different objects
```

Membership operators

```python
f = ['apple', 'banana', 'cherry', 'toyato', 'mazda']
print('honda' in f) #member false
print('mazda' in f) #member true
print('honda' not in f) #not member true
print('mazda' not in f) #not member false
```

Bitwise operators

```python
a = 10
b = 20
print( a & b) #bitwise and
print( a | b) #bitwise or
print( a ^ b) #bitwise xor
print( ~b) #bitwise not
print( ~a ) #bitwise not
print( a << b) #left shift
print( a >> b) #right shift
```

Condional statements

```python
#simple if
a = 10
b = 20
if a > b:
  print('a is greater than b')
else:
  print('b is greater than a')
```

```python
#nested if
a = 10
b = 20
if a > b:
  if a == 20:
    print('a is equal to 20')
  else:
    print('a is greater than b')
else:
  print('b is greater than a')
```

```python
#if-elif-else
a = 10
b = 20
if a > b:
  print('a is greater than b')
elif a == b:
  print('a is equal to b')
else:
  print('b is greater than a')
```

```python
#nested if-elif-else
a = 10
b = 20
if a > b:
  if a == 20:
    print('a is equal to 20')
  elif a == 30:
    print('a is equal to 30')
  else:
    print('a is greater than b')
else:
  print('b is greater than a')
```

Loops

```python
#for loop
for i in range(5):
  print(i)
```

```python
#for loop with element list
a = ['apple', 'banana', 'cherry', 'toyato', 'mazda']
for i in a:
  print(i)
```

```python
#for loop with element with condition
a = ['apple', 'banana', 'cherry', 'toyato', 'mazda']
for i in a:
  if i == 'cherry':
    continue
  print(i)
```

```python
#nested for loop
for i in range(5):
  for j in range(5):
    print(i,j)
```

```python
#while loop
i = 0
while i < 5:
  print(i)
  i += 1
```

```python
#nested while loop
i = 0
while i < 5:
  j = 0
  while j < 5:
    print(i,j)
    j += 1
  i += 1
```

```python
#while loop with break
i = 0
while i < 5:
  if i == 3:
    break
  i += 1
  print(i)
```

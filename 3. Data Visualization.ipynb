```python
b = sns.load_dataset('diamonds')
```

```python
b
```

```python
sns.distplot(b.depth)
```

```python
sns.boxplot(b.depth)
```

```python
sns.boxenplot(b.depth)
plt.show()
```

```python
sns.barplot(b.depth)
```

```python
c = b.groupby('cut')['cut'].count()
sns.barplot(x=c.index, y=c.values)
plt.title('Diamond Cut')
plt.xlabel('Cut Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.swarmplot(b.cut)
```

```python
d = sns.load_dataset('penguins')
```

```python
d
```

```python
e = d.groupby('species')['species'].count()
plt.pie(e)
plt.show()
```

```python
plt.pie(e, labels=e.index, autopct="%.2f", explode=[1, 0, 0], shadow=True)
plt.show()
```

```python
plt.pie(e, labels=e.index, autopct="%.2f", shadow=True)
plt.show()
```

```python
plt.pie(e, labels=e.index, autopct="%.2f", shadow=True, wedgeprops=dict(width=0.4))
plt.show()
```

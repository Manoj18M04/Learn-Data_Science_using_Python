Operators

```python
#Assignment Operators
print(15 + 8)
print(15 - 8)
print(15 * 8)
print(15 / 8)
print(15 // 8)
print(15 ** 8)
print(15 % 8)
```

```python
#Assignment Operator
a = 15
#a = a + 5
a += 5
print(a)
```

```python
a -= 10
print(a)
```

```python
a *= 2
print(a)
```

```python
a /= 5
print(a)
```

```python
a **= 5
print(a)
```

```python
a //= 5
print(a)
```

```python
#Comparison/Relational Operator
b = 5
c = 8
print(b == c)
print(b != c)
print(b < c)
print(b <= c)
print(b > c)
print(b >= c)
```

```python
#Logical Operator
d = 10
e = 15

print(not(d > 5 and e < 20))
print(not(d > 5 and e > 20))
print(not(d != 5 and e == 20))
```

```python
#Identity Operator
f = ['Abhishek', "Lordson"]
g = ['Rithik', 5, 6]
i = f
print(i is g)
print(f is g)
print(f is not g)
print(i is f)
```

```python
#Membership Operator
j = ['Abhishek', 'Lordson', 'Rithik', 5, 6]
print(7 in j)
print(7 not in j)
print('RIthik' in j)
print('Rithik' in j)
```

```python
#Bitwise Operator
& AND
| OR
^ XOR
~ NOT
<< Zero fill left shift
>> Signed right shift
```

```python
k = 10
l = 5

if l == k:
    print('Hii')
elif l < k:
    print('Heyy')
elif l != k:
    print('Hello')
else:
    print('Bye')
```

Loops

```python
#For Loop
for i in range(15):
    print(i)
```

```python
for i in range(3, 15):
    print(i)
```

```python
for i in range(3, 15, 2):
    print(i)
```

```python
m = ['Rithik', 'SECE', 'Abhishek', 'Lordson', 2, 'Abdulla', 8]
for z in m:
    print(z)
```

```python
m = ['Rithik', 'SECE', 'Abhishek', 'Lordson', 2, 'Abdulla', 8]
for z in m:
    print(z)
    if z == 'Abhishek':
        break
```

```python
m = ['Rithik', 'SECE', 'Abhishek', 'Lordson', 2, 'Abdulla', 8]
for z in m:
    if z == 'Abhishek':
        break
    print(z)
```

```python
#While Loop
n = 1
while n <= 10:
    n+=1
    print(n)
```

```python
n = 1
while n <= 10:
    print(n)
    n+=1
```
